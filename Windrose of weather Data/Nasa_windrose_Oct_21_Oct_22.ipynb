# Windrose Graph for 1 years Wind Data (Oct 2021 to Oct 2022)
## Location : Hub,Pakistan
--------

### 1. Loading the Dataset
------

```python
# Importing the pandas library
import pandas as pd

# Loading the dataset
Data=pd.read_excel("LST_Master.xlsx", sheet_name = "Oct-21-Oct-22", header=0)
```

```python
# Reviewing the dataset
Data.head()
```

### 2. Making the Windrose Graph
-------

```python
# Importing the windrose library
from windrose import WindroseAxes

# Extracting the wind_speed and wind_direction column from the main dataset
ws=Data.iloc[:,4]
wd=Data.iloc[:,5]

# Creating the graph
ax = WindroseAxes.from_ax()
ax.bar(wd, ws, opening=0.8, edgecolor='white',normed=True,bins=[0,0.5,2.1,3.6,5.7,8.8,11.1])

# Setting up the legends and Title of the graph
ax.set_legend()
ax.set_title("October 2021 to October 2022")
```
